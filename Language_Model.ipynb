# Loading of Dataset

Loading of Data as this is a small implementation therefore, i used only the plain text of **Alice in Wonderland** and building model on it.

```python
f = open("/content/drive/MyDrive/NLP/Alice in Wonderland.txt")
data = f.read()
print(data)
```

# Preprocessing of Dataset

```python
head, sep, tail = data.partition('*** START OF THE PROJECT GUTENBERG EBOOK ALICE IN WONDERLAND ***')
# print(sep)
head2, sep2, tail2 = tail.partition('*** END OF THE PROJECT GUTENBERG EBOOK ALICE IN WONDERLAND ***')
# print(tail2)
```

Now, head2 contain our data without any unncessary info.

```python
import re
txtx = '123,.,?{{""[adcv()]-&@#^*|-[]asda}/a/d/s}::sdsf44:;'
data = re.sub('[.,!?:;@#^*|(){}"ââ&ââ$%/]',"",txtx).replace("[","").replace("]","").replace("_","").replace("-","").
print(data)
data
```

```python
# Cleaning Text
import re
data_lower = head2.lower()
data_lower2 = re.sub('[.,!?:;@#^*|(){}"ââ&ââ$%/]',"",data_lower).replace("[","").replace("]","").replace("_","").replace("-","")
data_lower3 = ''.join(filter(lambda x: not x.isdigit(), data_lower2))
clean_data = data_lower3.replace("\n"," ").replace("\'","").replace("_","")
clean_data
```

```python
#Tokenization
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
token_list = word_tokenize(clean_data)
token_list
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(token_list)
word2indexdict = tokenizer.word_index
print('word dictionary', word2indexdict)
```

```python
index2worddict = {index: word for word, index in word2indexdict.items()}
print('reverse word dictionary', index2worddict)
```

```python
# creating trigram table
from nltk.util import ngrams
trigram = list(ngrams(token_list, 3))
trigram
```

```python
inputs = []
for i in trigram:
  inputs.append(list(i)[:2])
inputs
```

```python
x = []
for i in inputs:
  temp = []
  for word in i:
    temp.append(word2indexdict[word])
  x.append(temp)

x = np.array(x)
x
```

```python
outputs = []
for i in trigram:
  outputs.append(i[2])
outputs
```

```python
y = []

for i in outputs:
  y.append(word2indexdict[i])
y = np.array(y)
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Flatten, Dense
```

```python
vocab_size = len(word2indexdict) + 1
vocab_size
```

# Feed Forward Neural Network Embedding Layers
Model for prediciting the next word by training the model from this data text. Input is 2-word and Output is 1-word as per training.

```python
model = Sequential()

# 1. Embedding Layer
model.add(Embedding(input_dim=vocab_size + 1, output_dim=50, input_length=2))

# 2. Flatten Layer (turns 2 word vectors into one long vector)
model.add(Flatten())

# 3. Hidden Dense Layer
model.add(Dense(128, activation='relu'))

# 4. Output Layer (predict next word â total units = vocab size)
model.add(Dense(vocab_size, activation='softmax'))
```

```python
model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
model.fit(x_train, y_train, epochs=30, batch_size=32)
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {accuracy}")
```

```python
word1 = "wonderland"
word2 = "is"
x_input = np.array([[word2indexdict[word1], word2indexdict[word2]]])
prediction = model.predict(x_input)
predicted_index = np.argmax(prediction)
predicted_word = index2worddict[int(predicted_index)]
predicted_word
```
